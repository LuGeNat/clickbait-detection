```python
import json
from collections import Counter
```

```python
# samlpe ids from all tweets unanimously annotated 1.0 or 0.1

with open("truth_verification.jsonl", "r") as inf:
    truth_verification = [json.loads(x) for x in inf.readlines()]
with open("clickbait17-train-170331/truth.jsonl", "r") as inf:
    truth_verification_old = [json.loads(x) for x in inf.readlines()]

uni_clickbait = [int(i['id']) for i in truth_verification if Counter(i['truthJudgments'])[1.0] == 5 or (Counter(i['truthJudgments'])[1.0] == 4 and Counter(i['truthJudgments'])[0.6666666666])]
uni_clickbait_old = [int(i['id']) for i in truth_verification_old if Counter(i['truthJudgments'])[1.0] == 5 or (Counter(i['truthJudgments'])[1.0] == 4 and Counter(i['truthJudgments'])[0.6666666666])]
uni_no_clickbait = [int(i['id']) for i in truth_verification if Counter(i['truthJudgments'])[0.0] == 5]

print(len(uni_clickbait_old))
uni_clickbait = uni_clickbait + uni_clickbait_old
print(len(uni_clickbait))

with open("new_checks_clickbait.json", "w") as of:
    of.write(json.dumps(uni_clickbait))
        
with open("new_checks_no_clickbait.json", "w") as of:
    of.write(json.dumps(uni_no_clickbait[:len(uni_clickbait)]))
```
